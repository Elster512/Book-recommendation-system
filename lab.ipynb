## РЕКОМЕНДАЦИЯ ПО РЕЙТИНГУ ЮЗЕРОВ

```python
from surprise import Reader, Dataset, SVD
```

```python

reader = Reader(rating_scale=(1, 10))
data   = Dataset.load_from_df(final_rating[['userID','bookTitle','bookRating']], reader)
trainset = data.build_full_trainset()
model = SVD(n_factors=50, n_epochs=10, lr_all=0.005, reg_all= 0.2)
model.fit(trainset)
```

```python
def generate_recommendationsSVD(userID=277427, get_recommend =10):
    
    
    # predict rating for all pairs of users & items that are not in the trainset
    
    testset = trainset.build_anti_testset()
    predictions = model.test(testset)
    predictions_df = pd.DataFrame(predictions)

    # get the top get_recommend predictions for userID
    
    predictions_userID = predictions_df[predictions_df['uid'] == userID].\
                         sort_values(by="est", ascending = False).head(get_recommend)
    recommendations = []
    recommendations.append(list(predictions_userID['iid']))
    recommendations = recommendations[0]
    rec_books = sampleBooks[sampleBooks.bookTitle.isin(recommendations)]
    return(rec_books.to_dict('records'))
```

```python
from surprise.model_selection import cross_validate
cross_validate(model, data, measures=['RMSE', 'MAE'], cv=3, verbose=True)
```

```python
recommendationsSVD = generate_recommendationsSVD(userID=241980, get_recommend =30)
```

```python
def get_RecBooks_COSIAN(title):
    recBooks = []

    recommendations = pd.DataFrame(df.nlargest(11,title.lower())['bookTitle'])
    recommendations = recommendations[recommendations['bookTitle']!=title.lower()]
    recList = recommendations['bookTitle'].values.tolist()
    for i in range(len(recList)):
        recBooks.append(sampleBooks[sampleBooks['bookTitle'].str.lower() ==recList[i]].to_dict('records')[0])

    return recBooks
```

```python
from flask import Flask, request, jsonify
from flask_cors import CORS, cross_origin
app = Flask(__name__)
cors = CORS(app)
app.config['CORS_HEADERS'] = 'Content-Type'

```

```python

@app.route('/books')
@cross_origin()
def books():
    amount = len(booksDict)
    pages =math.ceil(amount/amount_of_books-1)
    page = request.args.get('page')
    if not page:
        return jsonify({'books':booksDict[0:amount_of_books],'pages':pages,'page':1}),200
    if int(page)*amount_of_books >len(booksDict):
        return jsonify({'error':'error'}),404
    return jsonify({'books':booksDict[(int(page))*amount_of_books:(int(page)+1)*amount_of_books],'pages':pages,'page':int(page)}),200



@app.route('/books/<id>')
@cross_origin()
def get_one_book(id):
    book = sampleBooks[sampleBooks.ISBN == str(id)].to_dict('records')
    if not book:
        return jsonify({'error':'not found','status':404}),200
    
    return jsonify({'body':book[0],'status':200}),200

@app.route('/get_r_user')
@cross_origin()
def get_r_user():
    random_user = random.randint(0,len(list_of_distinct_users))
    return jsonify({'body':{'userID':int(list_of_distinct_users[random_user])},'status':200}),200


@app.route('/rec_book')
@cross_origin()
def reccomendation_for_book():
    title = request.args.get('title')
    book =sampleBooks[sampleBooks.bookTitle == title].to_dict('records')
    
    if not title:
        return jsonify({'error':'not found','status':404}),200
    elif not book:
        return jsonify({'error':'not found','status':404}),200
    
    recBooks = get_RecBooks_COSIAN(title)
    
    return jsonify({'body':recBooks,'status':200}),200


@app.route('/rec_user')
@cross_origin()
def reccomendation_for_user():
    userID = request.args.get('userID')
    if not int(userID) in list_of_distinct_users:
        return jsonify({'error':'not found','status':404}),200
    
    recBooks = generate_recommendationsSVD(int(userID))
    print(recBooks)
    return jsonify({'body':recBooks,'status':200}),200
```
